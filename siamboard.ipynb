```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/122.0.0.0 Safari/537.36"
}

base_url = "https://siamboardgames.com/shop/all/?product-page="
```

```python
products = []

for page in range(1, 13):
    url = f"{base_url}{page}"
    response = requests.get(url, headers=headers)

    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        product_containers = soup.find_all('li', class_='product')

        for product in product_containers:
            # Get product name
            name_tag = product.find('h2', class_='woocommerce-loop-product__title')
            name = name_tag.get_text(strip=True) if name_tag else 'N/A'


            price_tag = product.find('span', class_='price')
            
            amount_tag = price_tag.find_all('span', class_='woocommerce-Price-amount')
            
            if(len(amount_tag) > 1):
                del_tag = product.find('del')
                if del_tag:
                    price = amount_tag[0].get_text(strip=True)
                    second_price = amount_tag[1].get_text(strip=True)
                    sale = True
                else:
                    price = amount_tag[0].get_text(strip=True)
                    second_price = amount_tag[1].get_text(strip=True)
                    sale = False
            else:
                price = amount_tag[0].get_text(strip=True) if len(amount_tag) > 0 else 'N/A'
                second_price = 'N/A'
                sale = False
            
            
            # Get product link
            link_tag = product.find('a', href=True)
            link = link_tag['href'] if link_tag else 'N/A'

            products.append({
                'name': name,
                'origin price': price,
                'add-on_price': 'N/A' if sale else second_price,
                'sale price': second_price if sale else 'N/A',
                'url': link,
                'sale': sale
            })
    else:
        print(f"Failed to fetch page {page}, status code: {response.status_code}")
```

```python
# Output results
for product in products:
    print(f"Name: {product['name']}, Price: {product['origin price']}, Add-on Price: {product['add-on_price']}, Sale Price: {product['sale price']},   URL: {product['url']}, Sale: {product['sale']}")
```

```python

full_proucts = []

for boardgame in products : 
    print(boardgame['name'], boardgame['url'])
    url = boardgame['url']
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        # 1. Small description (element with name="description")
        soup.find(attrs={"name": "description"}) is None
        small_description_element = soup.find(attrs={"name": "description"})
        if small_description_element:
            small_description = small_description_element.get("content")

        # 2. Image sources in gallery
        gallery_div = soup.find("div", class_="woocommerce-product-gallery--with-images")
        images_src = []
        if gallery_div:
            images = gallery_div.find_all("img")
            for img in images:
                images_src.append(img.get("src"))


        # 3. Full descriptions (all elements except first h2 under description panel)
        desc_panel = soup.find("div", class_="woocommerce-Tabs-panel--description")
        desc_list = []
        if desc_panel:
            skip_first_h2 = False
            for element in desc_panel.find_all(recursive=False):
                # Skip the first h2 only
                if element.name == "h2" and not skip_first_h2:
                    skip_first_h2 = True
                    continue
                desc_list.append(element.get_text(strip=True))

        # 4. Additional Information
        additional_panel = soup.find("div", class_="woocommerce-Tabs-panel--additional_information")
        additional_info = []
        if additional_panel:
            labels = additional_panel.find_all(class_="woocommerce-product-attributes-item__label")
            values = additional_panel.find_all(class_="woocommerce-product-attributes-item__value")
            for label, value in zip(labels, values):
                additional_info.append((label.get_text(strip=True), value.get_text(strip=True)))

        # 5. Equipment in the box (อุปกรณ์ในกล่อง)
        box_panel = soup.find("div", class_="woocommerce-Tabs-panel--อุปกรณ์ในกล่อง")
        box_contents = []
        if box_panel:
            for li in box_panel.find_all("li"):
                box_contents.append(li.get_text(strip=True))

        # 6. Sleeves (ซอง)
        sleeve_panel = soup.find("div", class_="woocommerce-Tabs-panel--ซอง")
        sleeve_contents = []
        if sleeve_panel:
            for li in sleeve_panel.find_all("li"):
                sleeve_contents.append(li.get_text(strip=True))

        full_proucts.append({
            'name': boardgame['name'],
            'price': boardgame['origin price'],
            'url': boardgame['url'],
            'small_description': small_description,
            'images_src': images_src,
            'full_description': desc_list,
            'additional_info': additional_info,
            'box_contents': box_contents,
            'sleeve_contents': sleeve_contents,
            'sale': boardgame['sale'],
            'sale price': boardgame['sale price'],
            'add-on_price': boardgame['add-on_price']
        })
    else:
        print(f"Failed to fetch product details for {boardgame['name']}, status code: {response.status_code}")
        
```

```python
with open('full_boardgames.csv', 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['name', 'price', 'url', 'small_description', 'images_src', 'full_description', 'additional_info', 'box_contents', 'sleeve_contents', 'sale', 'sale price', 'add-on_price']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    
    writer.writeheader()
    for product in full_proucts:
        writer.writerow(product)
```
